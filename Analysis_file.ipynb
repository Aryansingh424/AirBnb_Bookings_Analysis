```python
def total_rev():
    
    try:

        total_sum = AirBnb_df["price"].sum()
        return total_sum
    
    except Exception as e:
       
       print("Error occured")

```

```python
# Average of overall price that has been earned is :
```

```python
total_number_of_price = len(AirBnb_df["price"])
Sum_of_price = AirBnb_df["price"].sum()

def avrg():

    try:
        Average = Sum_of_price/total_number_of_price
        return Average
    
    except Exception as e:
        print("error occured")

    finally:
        print("Average is calculated successfuly")

Average_of_price = avrg()

print(f"Average of overall price is {Average_of_price}")

    

```
