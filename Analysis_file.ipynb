```python
# Dataset First Look

AirBnb_df = pd.read_csv(dir_path)
AirBnb_df
```

```python
# Dataset Duplicate Value Count

def check_duplicates():

    try:
        dup_count = AirBnb_df.duplicated().sum()

        return dup_count 

    except Exception as e:

        print("please give relatable data")
 
```

```python
how_many_duplicate_rows = check_duplicates()

print(f'The total number of duplicate rows excluding first occurence {how_many_duplicate_rows}')
```

```python
# (The total number of duplicate rows excluding first occurence 0 ) 
# This output states that there are no duplicates rows in the given data 
```

```python
# counting null values for each column separately

null_values = AirBnb_df.isna().sum()
print(null_values)
```

```python
# counting total number of null values in the data frame 

null_values = null_values.sum()
print(null_values)
```

```python
# visualizing the missing Values 


plt.figure(figsize = (10,50))
AirBnb_df.isna().sum().plot(kind='bar')
plt.title('Null Values Visualization')
plt.xlabel('Columns of DataFrame')
plt.ylabel('Numbers of Rows')
plt.xticks(rotation = 60)
plt.ylim(0, len(AirBnb_df))

# figure size column is 50 because there are very few null values in name and host name also which might not be visible at 10, 5 view
```
